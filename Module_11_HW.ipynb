A histogram displays continuous data. This frequency helps a user spot trends.

```
#histogram set 1
df.hist(figsize=(10, 10), bins=30, xlabelsize=10, ylabelsize=10)
```

Heatmaps display popular data in colors, showing relation in a visual representation. 

```
#heatmap
corr = df.corr()
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(data=corr, square=True, annot=True, cbar=True, ax = ax)
```

```
#heatmap
corr = df2.corr()
fig, ax = plt.subplots(figsize=(9,9))
sns.heatmap(data=corr, square=True, annot=True, cbar=True, ax = ax)
```

This chart shows normalization and makes the features of the data display more consistent with each other and allow predictability.

```
normalized_df=(df-df.mean())/df.std()
for col in df:
  sns.kdeplot(normalized_df[col], shade=True)
```

```
normalized_df2=(df2-df2.mean())/df2.std()
for col in df2:
  sns.kdeplot(normalized_df2[col], shade=True)
```
