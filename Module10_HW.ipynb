**LISTS vs. TUPLES**
Both lists and tuples store one or more values in a specific order. 
*Differences*
Lists:
- are surrounded by square brackets
- can be changed after creation
- more built in functions than tuples

Tuples:
- are surrounded by parenthesis
- immutable nature and can't be changed or modified after it is created
- smaller in size
